# Numpy Index

## Indexing 1D Array

```python
import numpy as np
```

```python
arr = np.arange(0, 11)
```

```python
arr[8]
```

```python
arr[1:5]
```

```python
arr[:5]
```

```python
arr[5:]
```

## Broadcasting

```python
arr[0:5] = 100
```

```python
arr
```

```python
arr = np.arange(0, 11)
```

```python
slice_of_arr = arr[0:5]
```

```python
slice_of_arr
```

```python
slice_of_arr[:] = 99
```

```python
slice_of_arr
```

```python
arr           ## it also changed Parent Array
```

```python
slice_of_arr = arr.copy()
```

```python
slice_of_arr[:] = 100
```

```python
slice_of_arr
```

```python
arr
```

## Indexing 2D arrays

```python
arr2d = np.array([[5, 10, 15], [20, 25, 30], [35, 40, 45]])
```

```python
arr2d
```

```python
arr2d.shape
```

```python
arr2d[0]
```

```python
arr2d[-1]
```

```python
arr2d[1][1]
```

```python
arr2d[1, 1]
```

```python
arr2d[:2]
```

## Conditional Selection

```python
arr = np.arange(1, 11)
```

```python
arr > 4
```

```python
boolean_arr = arr > 4
arr[boolean_arr]
```

```python
arr[arr > 4]
```
