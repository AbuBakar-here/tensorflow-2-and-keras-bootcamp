## Missing Data

```python
import pandas as pd
import numpy as np
```

```python
df = pd.DataFrame({'A': [1, 2, np.nan, 4],
                   'B': [5, np.nan, np.nan, 8],
                   'C': [10, 20, 30, 40]})
```

```python
df
```

## Drop Values

```python
df.dropna()
```

```python
df.dropna(axis = 1)
```

```python
df.dropna(axis = 1, thresh = 3)
```

## Fill in Values

```python
df.fillna(value = 'FILL VALUE')
```

```python
df.fillna(value = 0)
```

```python
df['A'] = df['A'].fillna(0)
```

```python
df
```

```python
df['B'] = df.fillna(df['B'].mean())
```

```python
df = pd.DataFrame({'A': [1, 2, np.nan, 4],
                   'B': [5, np.nan, np.nan, 8],
                   'C': [10, 20, 30, 40]})
df.fillna(df.mean())
```
