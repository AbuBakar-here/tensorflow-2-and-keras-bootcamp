# Pandas Series

```python
import numpy as np
import pandas as pd
```

## Series

```python
labels = ['a', 'b', 'c']

mylist = [10, 20, 30]

arr = np.array([10, 20, 30])

d = {'a' : 10, 'b' : 20, 'c' : 30}
```

```python
pd.Series(data = mylist)
```

```python
pd.Series(data = mylist, index = labels)
```

```python
pd.Series(arr, labels)
```

```python
pd.Series(d)
```

## Indexing

```python
salesQ1 = pd.Series(data = [250, 450, 200, 150], index = ['USA', 'CHINA', 'India', 'Brazil'])
```

```python
salesQ1
```

```python
salesQ2 = pd.Series(data = [260, 500, 210, 100], index = ['USA', 'CHINA', 'India', 'Japan'])
```

```python
salesQ2
```

```python
salesQ2['CHINA']
```

```python
salesQ2[0]
```

## Operations

```python
salesQ1 + salesQ2
```

```python
salesQ1 - salesQ2
```
