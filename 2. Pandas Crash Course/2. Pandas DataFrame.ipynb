# Pandas DataFrame

```python
import numpy as np
import pandas as pd
```

## DataFrame

```python
columns = ['W', 'X', 'Y', 'Z']

index = ['a', 'b', 'c', 'd', 'e']
```

```python
np.random.seed(42)

data = np.random.randint(-100, 100, size = (5, 4))
```

```python
data
```

```python
df = pd.DataFrame(data = data, index = index, columns = columns)
```

```python
df
```

```python
df['W']
```

```python
type(df['W'])
```

```python
df[['W', 'X']]
```

```python
df['new'] = df['W'] + df['Y']
```
