# Matplolib Basics

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
%matplotlib inline
```

```python
x = [0, 1, 2]
y = [100, 200, 300]
```

## Line Plot

```python
plt.plot(x, y)
```

## Scatter Plot

```python
housing = pd.DataFrame({'rooms': [1, 1, 2, 2, 2, 3, 3, 3],
                        'price': [100, 120, 190, 200, 230, 310, 330, 305]})
```

```python
housing
```

```python
plt.scatter(housing['rooms'], housing['price'])
```

## Title, Legends and etc

```python
plt.plot(x, y)
plt.title("Title")
plt.xlabel("x label")
plt.ylabel("y label")
```

```python
plt.plot(x, y)
plt.title("Title")
plt.xlabel("x label")
plt.ylabel("y label");
```

```python
plt.plot(x, y, color = 'red', marker = '*', markersize = 15, linestyle = '--')

plt.xlim(0, 2)
plt.ylim(100, 300)

plt.title("Title")
plt.xlabel("x label")
plt.ylabel("y label");
```
