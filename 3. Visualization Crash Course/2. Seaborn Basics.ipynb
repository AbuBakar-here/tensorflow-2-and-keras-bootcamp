# Seaborn Basics

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
from google.colab import drive
drive.mount("/drive")
```

```python
sns.mpl.rcParams["figure.figsize"] = [12, 8]
sns.set_style()
sns.set_theme()
```

```python
data = pd.read_csv("/drive/My Drive/Colab Notebooks/Deep Learning Course (Jose Portilia)/3. Visualization Course/heart.csv")
```

```python
data.head()
```

## Distribution Plot

```python
sns.displot(data['age'], bins = 20)
```

## Count Plot

```python
sns.countplot(x = 'sex', data = data)
```

```python
sns.countplot(x = 'target', data = data)
```

```python
sns.countplot(x = 'cp', data = data, hue = 'sex')
```

## Box Plot

```python
sns.boxplot(x = 'sex', y = 'age', data = data)
```

```python
sns.boxplot(x = 'target', y = 'thalach', data = data, hue = 'sex')
```

## Scatter Plot

```python
sns.scatterplot(x = 'chol', y = 'trestbps', data = data, hue = 'sex', palette = 'Dark2')
```
